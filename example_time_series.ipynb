```python
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
```

```python
def parser(x):
    #convert input data to a readable format 190x-xx
	return datetime.strptime('190'+x, '%Y-%m')
```

```python
series = read_csv('shampoo-sales.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
print (type(series))
```

```python
print(series.head())
series.plot()
pyplot.show()
```

```python
#from pandas.tools.plotting import autocorrelation_plot
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(series)
pyplot.show()
```

```python
import pandas
import numpy
pyplot.gcf().clear()#clearing neccessary due to previous plot accumulating % for performance.
pyplot.figure()
data = pandas.Series(0.7 * numpy.random.rand(1000) +
    0.3 * numpy.sin(numpy.linspace(-9 * numpy.pi, 9 * numpy.pi, num=1000)))
autocorrelation_plot(data)
pyplot.show()
```
